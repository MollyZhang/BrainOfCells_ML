```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh import palettes

%matplotlib inline
output_notebook()
```

### Plot first two components of PCA in all data

```python
df = pd.read_csv("./data_v4/ExpressionMatrix_PCA_0.4_variance.csv", index_col="sample")
```

```python
p = figure(plot_width=800, plot_height=700)
p.circle(df["0"], df["1"], size=3)
show(p)
```

```python
dataset_df = pd.read_csv("./data_v4/dataset_id.csv", index_col="sample_id")
```

```python
labeled_df = df.join(dataset_df)
datasets = list(set(dataset_df["#data_set_id"]))
```

```python
p = figure(plot_width=800, plot_height=700)
colors = palettes.Category20[20]
for i, dataset in enumerate(datasets):
    subset_df = labeled_df[labeled_df["#data_set_id"] == dataset]
    p.circle(subset_df["0"], subset_df["1"], size=3, color=colors[i], legend=dataset)

p.legend.click_policy="hide"
show(p)
```
